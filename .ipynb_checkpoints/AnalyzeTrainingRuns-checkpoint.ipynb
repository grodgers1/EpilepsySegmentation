### Import libraries and set up directories + runs to compare

```python
import os
import pandas as pd
import glob

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Directories
BasePath = '/scicore/home/muelleb/panile54/epilepsy_segmentation/UNET_v6/'
CheckpointsPath = 'checkpoints1'

BaseDir = os.path.join(BasePath,CheckpointsPath)
```

```python
# which training runs to analyze/compare?
Runs = ['default','depth3','depth4','dropout1','dropout2','flips','noise1','randaffine1','randaffine2','width16','width8']
Nicknames = ['default','depth3','depth4','dropout1','dropout2','flips','noise1','randaffine1','randaffine2','width16','width8']

assert len(Runs) == len(Nicknames)
```

### Collect metadata about runs from log files

```python
# collect metadata
df_list, df2_list = [],[]
for idx, ThisRun in enumerate(Runs):
    df = pd.read_csv(os.path.join(BaseDir,ThisRun,'network_parameters.txt'),sep=':\t',header=None,engine='python').T
    new_header = df.iloc[0]
    df = df[1:]
    df.columns = new_header
    df_list.append(df)
    
    df2 = pd.read_csv(os.path.join(BaseDir,ThisRun,'logs.csv'))
    df2['model_name'] = ThisRun
    df2['Nickname'] = Nicknames[idx]
    
    df2_list.append(df2)
    
ModelMetadata = pd.concat(df_list,axis=0).reset_index().drop(columns=['index'])
ModelMetadata = pd.concat([pd.DataFrame(data={'Nicknames': Nicknames}),ModelMetadata],axis=1)
ModelMetadata[['Nicknames','model_name','loss','total_learned_parameters']]

LearningCurves = pd.concat(df2_list,axis=0).reset_index()
```

```python
LearningCurves
```

### Collect information about training and validation set

```python
# find number of images in training set from each sample
TrainDir = os.path.join(BasePath,'train1')
ValDir = os.path.join(BasePath,'val1')

ImList = os.listdir(os.path.join(TrainDir,'images'))
SampleNames = [x[:-8] for x in ImList]
SampleNames = set(SampleNames) # removes duplicates
SampleNames = list(SampleNames) # list of samples

print(f'There are {len(ImList)} manually labelled images from {len(SampleNames)} samples')

print('Samples:')
print(SampleNames)

numImsTrainList = []
numImsValList = []
for thisSample in SampleNames:
    theseImsTrain = glob.glob(os.path.join(TrainDir,'images',thisSample + '*.png'))
    numImsTrain = len(theseImsTrain)
    numImsTrainList.append(numImsTrain)
    
    theseImsVal = glob.glob(os.path.join(ValDir,'images',thisSample + '*.png'))
    numImsVal = len(theseImsVal)
    numImsValList.append(numImsVal)

SampleNames[2] = 'PCB1KA21dlh'
SampleNames[6] = 'PCB2KA21drh'
Original_SampleNames = SampleNames
d = {'samplename':SampleNames,'NumberTrainingIms':numImsTrainList,'NumberValIms':numImsValList}
df = pd.DataFrame(data=d)
df
```

### Plot and save the learning curves for each run

```python
for idx, ThisRun in enumerate(Runs):
    TheseCurves = LearningCurves[LearningCurves["model_name"] == ThisRun]
    
    fig, axs = plt.subplots(1,2, sharex=True, sharey=False, constrained_layout=True)
    fig.set_size_inches(12,5)

    axs[0].semilogy(TheseCurves['epoch'],TheseCurves['train_loss'],'b.')
    axs[0].set_title('train_loss')
    axs[0].set_xlabel('epoch')
    
    axs[1].semilogy(TheseCurves['epoch'],TheseCurves['val_loss'],'r.')
    axs[1].set_title('val_loss')
    axs[1].set_ylim(axs[0].get_ylim())
    axs[1].set_xlabel('epoch')

    #axs[1,0].plot(TheseCurves['epoch'],TheseCurves['Dice'],'r.')
    #axs[1,0].set_title('Dice')
    #axs[1,0].set_xlabel('epoch')

    #axs[1,1].plot(TheseCurves['epoch'],TheseCurves['Jaccard'],'r.',)
    #axs[1,1].set_title('Jaccard')
    #axs[1,1].set_ylim([0,1])
    #axs[1,1].set_xlabel('epoch')

    fig.suptitle('Model: ' + ThisRun);
    
    fname = os.path.join(BaseDir,ThisRun,'LearningCurves_' + ThisRun + '.png');
    plt.savefig(fname)
```

### Find best model for each based on learning curves
i.e. that with minimum loss on validation set

```python
BestModels = []
for idx, ThisRun in enumerate(Runs):
    TheseCurves = LearningCurves[LearningCurves["model_name"] == ThisRun]
    
    BestModel = TheseCurves.loc[TheseCurves['val_loss'] == TheseCurves['val_loss'].min()]    
    BestModels.append(BestModel)

BestModels = pd.concat(BestModels,axis=0).reset_index().drop(columns=['index']).set_index('model_name').drop("noise1")
BestModels
```

```python
# Plot similarity metrics from learning curves
fig, axs = plt.subplots(2,2, sharex=True, constrained_layout=True)
fig.set_size_inches(12,8)

a = BestModels.plot.scatter(x='Nickname',y='val_loss', ylabel='val_loss', legend=False, ax=axs[0,0])

b = BestModels.plot.scatter(x='Nickname',y='F1', ylabel='F1', legend=False, ax=axs[0,1])

c = BestModels.plot.scatter(x='Nickname',y='Dice', ylabel='Dice', legend=False, ax=axs[1,0])

d = BestModels.plot.scatter(x='Nickname',y='Jaccard', ylabel='Jaccard', legend=False, ax=axs[1,1])

axs[1,0].set_xlabel(None)
axs[1,1].set_xlabel(None)

axs[1,0].tick_params(axis='x',labelrotation=60)
axs[1,1].tick_params(axis='x',labelrotation=60)

fname = os.path.join(BaseDir,'Ablation_validationsetsummary1.png');
plt.savefig(fname)
```

## Detailed analysis of network performance on validation set
Make sure to run predict_val_channels.py on all datasets first

```python
df_list = []
for idx, ThisRun in enumerate(Runs):
    if not (ThisRun == 'noise1'):
        df = pd.read_csv(os.path.join(BaseDir,ThisRun,'best_model','metrics.csv'))
        df.drop(columns=df.columns[0],axis=1,inplace=True)
        Filenames = pd.Series(df['filename'])
        SampleNames = Filenames.str.split(n=1,pat='_',expand=True)[0].to_frame('SampleName')
        ImageNumbers = Filenames.str.rsplit(n=1,pat='_',expand=True)[1].to_frame('ImageNumber').astype(int)
        
        df = pd.concat([SampleNames,ImageNumbers,df],axis=1)
        df['model_name'] = ThisRun
        df['Nickname'] = Nicknames[idx]

        df_list.append(df)
        #print(ThisRun)

Metrics = pd.concat(df_list,axis=0).reset_index()
Metrics = Metrics.drop(Metrics[Metrics['filename']=='PCB1KA21dlh_offaxis_try3_287'].index)
SampleNames = pd.unique(Metrics['SampleName'])
```
