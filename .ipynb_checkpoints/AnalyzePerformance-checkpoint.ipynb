Import libraries

```python
import os
import pandas as pd
import numpy as np
import glob

import matplotlib.pyplot as plt
%matplotlib inline
```

Analyzing a single run:

```python
# Directories
BasePath = '/scicore/home/muelleb/panile54/epilepsy_segmentation/UNET_v5/'
CheckpointsPath = 'checkpoints0'

BaseDir = os.path.join(BasePath,CheckpointsPath)
```

```python
# Which model to analyze
ThisRun = 'alldefault'
```

```python
# Load metadata about run
NetworkParams = pd.read_csv(os.path.join(BaseDir,ThisRun,'network_parameters.txt'),sep=':\t',header=None,engine='python',names=["Name","Value"])
```

```python
NumParameters = NetworkParams.loc[NetworkParams["Name"]=='total_learned_parameters']["Value"].astype(str).astype(int).to_numpy()
print(f'Number of learned parameters: {NumParameters}')
```

```python
# Look at learning curve
TheseCurves = pd.read_csv(os.path.join(BaseDir,ThisRun,'logs.csv'))
TheseCurves.head()
```

```python
# Plot curves
fig, axs = plt.subplots(2,2, sharex=True, sharey=False, constrained_layout=True)
fig.set_size_inches(12,12)

axs[0,0].semilogy(TheseCurves['epoch'],TheseCurves['train_loss'],'b.')
axs[0,0].set_title('train_loss')

axs[0,1].plot(TheseCurves['epoch'],TheseCurves['jaccard'],'r.')
axs[0,1].set_title('val_jaccard')
axs[0,1].set_ylim([0,1])

axs[1,0].semilogy(TheseCurves['epoch'],TheseCurves['val_loss'],'r.')
axs[1,0].set_title('val_loss')
axs[1,0].set_xlabel('epoch')

axs[1,1].plot(TheseCurves['epoch'],TheseCurves['f1'],'r.',)
axs[1,1].set_title('val_f1')
axs[1,1].set_ylim([0,1])
axs[1,1].set_xlabel('epoch')

fig.suptitle('Model: ' + ThisRun)
```

```python
# Start by looking at a single checkpoint
TheseMetrics = pd.read_csv(os.path.join(BaseDir,ThisRun,'metrics.csv'))
#TheseMetrics.sort_values(by='samplename').head()
TheseMetrics.head()
```

```python
SampleNames = pd.unique(TheseMetrics['samplename'])
print(SampleNames)
```

```python
TheseMetrics.loc[TheseMetrics['samplename']==SampleNames[0]]
```

```python
fig, axs = plt.subplots(3, sharex=True, constrained_layout=True)
fig.set_size_inches(7,12)

a = TheseMetrics.boxplot(column='crossentropy',by='samplename',ax=axs[0])
a.set_xlabel('')
a.set_title('')
a.set_ylabel('cross entropy')
a.set_ylim([0,1])

b = TheseMetrics.boxplot(column='dice',by='samplename',ax=axs[1])
b.set_xlabel('')
b.set_title('')
b.set_ylabel('dice')
b.set_ylim([0,1])

c = TheseMetrics.boxplot(column='jaccard',by='samplename',ax=axs[2])
c.set_xlabel('')
c.set_title('')
c.set_ylabel('jaccard')
c.set_ylim([0,1])

plt.xticks(rotation=90)
fig.suptitle('');
```

```python
TheseMetrics.describe()
```

```python
# find number of images in training set from each sample
TrainDir = os.path.join(BasePath,'train0')

numImsList = []
for thisSample in SampleNames:
    theseIms = glob.glob(os.path.join(TrainDir,'images',thisSample + '*.png'))
    
    numIms = len(theseIms)
    numImsList.append(numIms)

d = {'samplename':SampleNames,'numIms':numImsList}
df = pd.DataFrame(data=d)
MetricList = TheseMetrics.groupby('samplename').median().reset_index()[['samplename','jaccard','dice']]
df2 = pd.merge(df,MetricList)

df2
```

```python
fig, axs = plt.subplots(2, sharex=True, constrained_layout=True)
fig.set_size_inches(5,5)

a = df2.plot.scatter(x='numIms',y='jaccard',ax=axs[0])
a.set_ylabel('Validation Jaccard')
b = df2.plot.scatter(x='numIms',y='dice',ax=axs[1])
b.set_ylabel('Validation Dice')
b.set_xlabel('Number of images in training set');
```
